```python
from selenium.webdriver.support import expected_conditions as Conditions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium import webdriver
from time import sleep

URL = 'https://tproger.ru/quiz/real-programmer/'
questions_count = 12
WAIT_TIME = 1.5


# Функция для задержки
def delay():
	driver.implicitly_wait(WAIT_TIME)
	sleep(WAIT_TIME)


# Ждет прогрузки, прокручивает, чтобы форма была видима, сохраняет элементы с ответами
def prepare_page():
	global form, form_source, answer_inputs
	form = wait.until(Conditions.visibility_of_element_located((By.CLASS_NAME, 'tp-quiz__container-quiz')))
	driver.execute_script('arguments[0].scrollIntoView();', form)
	answer_inputs = get_answer_inputs()


# Получает текст из комментария об ответе
def get_answer_comment() -> str:
	return wait.until(Conditions.visibility_of_element_located((By.CLASS_NAME, 'tp-quiz-answer-explanation__text'))).text


# Получаем элементы вариантов ответа
def get_answer_inputs():
	global form
	return form.find_elements(By.CSS_SELECTOR, '.tp-answer__radio')


# Пропускаем `count` вопросов
def skip_questions(count: int):
	global answer_inputs
	for _ in range(count):
		answer_inputs = get_answer_inputs()
		answer_inputs[0].click()
		delay()
		driver.find_element(By.CLASS_NAME, 'tp-button--primary').click()
		delay()


# Создаем драйвер, открываем сайт, инициализируем класс для ожидания
driver = webdriver.Chrome()
driver.get(URL)
wait = WebDriverWait(driver, 10)

prepare_page()


# Переменная для хранения элементов - ответов на вопросы
answer_inputs = get_answer_inputs()

# Идем по всем вопросам что есть (их число записано в программе - 12)
for question_number in range(questions_count):
	print(f'Question {question_number + 1}')

	# Прокликаем вопросы, которые уже нас не интересуют, то есть все вопросы до текущего
	skip_questions(question_number)
	# print('Skipped previous questions')

	# Выбираем первый вариант ответа - не знаем, верен он или нет
	answer_inputs = get_answer_inputs()
	current_answer = answer_inputs[0]
	current_answer.click()

	delay()

	right_answer_id = 0
	# Смотрим, какой ответ выделен зеленым и соответственно получаем его номер (нумерация с 0)
	# определяем цвет по классу CSS
	for answer in answer_inputs:
		if 'tp-answer--result-true' in answer.find_element(By.XPATH, './../..').get_attribute('class'):
			# Зеленый, значит его номер и есть номер верного ответа
			break
		else:
			# Не зеленый, значит смотрим следующий
			right_answer_id += 1
	# print(f'Got right answer id for question {question_number + 1}: {right_answer_id}')

	# Получаем текст правильного ответа
	right_answer_text = answer_inputs[right_answer_id].find_element(By.XPATH, '../div[1]/span[1]').text

	# Если не повезло и первый ответ не оказался верным
	if right_answer_id != 0:
		# print(f'Going to question {question_number + 1} again')
		# Обновляем страницу, заново ее подготавливаем
		driver.refresh()
		prepare_page()
		delay()
		# Заново пропускаем все вопросы
		skip_questions(question_number)
		delay()
		# Получаем ответы и кликаем, но уже на заведомо верный
		answer_inputs = get_answer_inputs()
		answer_inputs[right_answer_id].click()
		delay()
	else:
		# Тут нам повезло и мы уже нажали на верный ответ
		# print(f'Got right answer from first attempt')
		pass

	# Получаем текст вопроса и выводим его
	question_text = wait.until(Conditions.visibility_of_element_located((By.CLASS_NAME, 'tp-quiz-question__text'))).text
	print(question_text)

	
	# Получаем варианты ответа и для каждого из них получаем текст
	# текст лежит в соседнем элементе, так что получаем его по XPATHw
	answer_inputs = get_answer_inputs()
	for answer in answer_inputs:
		print(answer.find_element(By.XPATH, '../div[1]/span[1]').text)
	
	# Получаем комментарий о правильном ответе
	print(get_answer_comment())
	print(f'Right_answer - {right_answer_text}\n')

	# Обновляем страницу и подготавливаем заново
	driver.refresh()
	prepare_page()

# print('Quitting')
# Выходим из браузера
driver.quit()
```
